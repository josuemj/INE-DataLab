```python
import pandas as pd

# Diccionario de transformación
fall_les_dict = {
    1: "Fallecido",
    2: "Lesionado",
    3: "Ignorado"
}

# Asegurarse de que los valores sean numéricos si vienen como string
data["fall_les"] = pd.to_numeric(data["fall_les"], errors="coerce")

# Aplicar la transformación
data["fall_les"] = data["fall_les"].map(fall_les_dict)

# Mostrar los primeros resultados
data.head()
```

```python
tipo_veh_dict = {
    1: "Automóvil",
    2: "Camioneta sport o blazer",
    3: "Pick up",
    4: "Motocicleta",
    5: "Camión",
    6: "Cabezal",
    7: "Bus extraurbano",
    8: "Jeep",
    9: "Microbús",
    10: "Taxi",
    11: "Panel",
    12: "Bus urbano",
    13: "Tractor",
    14: "Moto taxi",
    15: "Furgón",
    16: "Grúa",
    17: "Bus escolar",
    18: "Bicicleta",
    19: "Avioneta",
    20: "Montacargas",
    21: "Bus militar",
    22: "Cuatrimoto",
    23: "Furgoneta",
    24: "Motos acuáticas",
    99: "Ignorado"
}

# Intentar convertir a número y transformar solo si es válido
data["tipo_veh_num"] = pd.to_numeric(data["tipo_veh"], errors="coerce")
data.loc[data["tipo_veh_num"].notna(), "tipo_veh"] = data["tipo_veh_num"].map(tipo_veh_dict)

# Eliminar la columna auxiliar
data.drop(columns="tipo_veh_num", inplace=True)

data
```
